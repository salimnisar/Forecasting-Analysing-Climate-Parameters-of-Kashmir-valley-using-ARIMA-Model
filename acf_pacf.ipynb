```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
data=pd.read_csv('./datasets/ceda/tmp/tmp5.csv',index_col='DATE',parse_dates=['DATE'])
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,5))
plot_acf(data['TEMPERATURE'],ax=ax[0],lags=20)
plot_pacf(data['TEMPERATURE'],ax=ax[1],lags=75)
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(18,5))
data['diff1']=data['TEMPERATURE'].diff()
plot_acf(data['diff1'].dropna(),ax=ax[0],lags=20)
plot_pacf(data['diff1'].dropna(),ax=ax[1],lags=50)
plt.show()
```
