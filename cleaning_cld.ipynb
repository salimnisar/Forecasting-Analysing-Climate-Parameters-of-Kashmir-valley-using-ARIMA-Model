```python
data=pd.read_csv('./datasets/ceda/cld/cld.csv',header=None)
data.drop(data.index[data.isnull().any(axis=1)],0,inplace=True)
data.drop(data.columns[0],axis=1,inplace=True)
data.to_csv('./datasets/ceda/cld/cld2.csv')
```

```python
data=pd.read_csv('./datasets/ceda/cld/cld5.csv',index_col=0, parse_dates=True)
data.reset_index(inplace=True)
data['MONTH'] = data['DATE'].dt.month
data['DATE'] = data['DATE'].dt.year
data.rename(columns={'DATE':'YEAR'},inplace=True)
data=data[['YEAR','MONTH','TEMPERATURE']]
data=pd.DataFrame(data)
data.head()
```
